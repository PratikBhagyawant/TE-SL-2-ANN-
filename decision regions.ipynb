Problem Statement:
With a suitable example demonstrate the perceptron learning law with its decision regions using
python. Give the output in graphical form.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step activation function
step_function = lambda x:1 if x > 0 else 0

# AND Gate Training Data
training_data = [
    {'input': [0, 0, 1], 'label': 0},  # bias = 1
    {'input': [0, 1, 1], 'label': 0},
    {'input': [1, 0, 1], 'label': 0},
    {'input': [1, 1, 1], 'label': 1}
]

# Initialize weights (2 inputs + bias)
weights = np.random.rand(3)
learning_rate = 0.1

# Training the Perceptron
for epoch in range(10):  # Multiple epochs for convergence
    for data in training_data:
        x = np.array(data['input'])
        y = data['label']
        output = step_function(np.dot(x, weights))
        error = y - output
        weights += learning_rate * error * x

print("Final weights:", weights)

# Decision Boundary Plotting
x_vals = np.linspace(-0.1, 1.1, 100)
y_vals = -(weights[0] * x_vals + weights[2]) / weights[1]

# Plotting
for point in training_data:
    color = 'blue' if point['label'] == 0 else 'red'
    plt.scatter(point['input'][0], point['input'][1], color=color)

plt.plot(x_vals, y_vals, '--k', label='Decision Boundary')
plt.show()
```
