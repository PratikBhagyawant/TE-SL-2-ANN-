Problem Statement :\
Write a Python program to plot a few activation functions that are being used in neural networks.\

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#consider unipolar where point is from 0 to 1
x = np.linspace(0,10,100)
```

```python
#Linear Funtion
y = x
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.grid(True)
plt.show()
```

```python
#Sigmoid Funtion
y = 1/(1+np.exp(-x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid Function')
plt.grid(True)
plt.show()
```

```python
#Tanh function
y = 2/(1+np.exp(-2*x)) -1
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tanh Function')
plt.grid(True)
plt.show()
```

```python
#Relu Funtion
y=np.maximum(0,x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function')
plt.grid(True)
plt.show()
```

```python
#SoftSign Function
y = x/(1+np.abs(x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('SoftSign Functions')
plt.grid(True)
plt.show()
```

```python
#Swish Function
y = x/(1+np.exp(-x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Swish Functions')
plt.grid(True)
plt.show()
```

```python
#Tanh Shrink function
y= x - np.tanh(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tanh Shrink Functions')
plt.grid(True)
plt.show()
```

```python
#Soft plus function
y = np.log(1+np.exp(x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Soft Plus Functions')
plt.grid(True)
plt.show()
```

```python
#Mish Function
y = x*np.tanh(np.log(1+np.exp(x)))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mish Functions')
plt.grid(True)
plt.show()
```

```python
#leaky Relu Function
y = np.maximum(x,x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Leaky ReLU Functions')
plt.grid(True)
plt.show()
```
